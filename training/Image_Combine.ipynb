```python
import os
import random
import cv2
import numpy as np

# Function to combine images from multiple folders into one
def combine_random_coins_from_folders(folders, output_dir, min_coins=1, max_coins=8, img_size=(800, 800), num_images=1500):
    # Check if the output directory exists, and create it if it doesn't
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Loop to generate the required number of images
    for i in range(num_images):
        # Create a blank canvas to place coins on
        canvas = np.ones((img_size[0], img_size[1], 3), dtype=np.uint8) * 255  # White background
        
        # List to track positions of placed coins
        placed_coins = []

        # Randomly select the number of coins to add
        num_coins = random.randint(min_coins, max_coins)

        # Loop to add coins to the canvas
        for _ in range(num_coins):
            # Randomly choose a folder
            folder = random.choice(folders)
            
            # Randomly select an image from the folder
            coin_images = os.listdir(folder)
            coin_image = random.choice(coin_images)
            
            # Load the selected image
            coin_path = os.path.join(folder, coin_image)
            coin = cv2.imread(coin_path)

            # Resize coin to fit in canvas (optional resizing)
            coin = cv2.resize(coin, (random.randint(50, 150), random.randint(50, 150)))

            # Try to find a valid position to place the coin (without overlap)
            placed = False
            for _ in range(100):  # Try up to 100 random positions before giving up
                # Randomly choose the position to place the coin
                max_x = canvas.shape[1] - coin.shape[1]
                max_y = canvas.shape[0] - coin.shape[0]
                x = random.randint(0, max_x)
                y = random.randint(0, max_y)

                # Check if the coin overlaps with any already placed coin
                overlap = False
                for (px, py, pw, ph) in placed_coins:
                    if (x < px + pw and x + coin.shape[1] > px and 
                        y < py + ph and y + coin.shape[0] > py):
                        overlap = True
                        break
                
                # If no overlap, place the coin
                if not overlap:
                    canvas[y:y + coin.shape[0], x:x + coin.shape[1]] = coin
                    placed_coins.append((x, y, coin.shape[1], coin.shape[0]))
                    placed = True
                    break

            # If no valid position found after 100 tries, skip the coin (optional)
            if not placed:
                print(f"Could not place coin {_+1} in image {i+1}, skipping.")

        # Save the resulting image
        output_image_path = os.path.join(output_dir, f"combined_image_{i+1}.jpg")
        cv2.imwrite(output_image_path, canvas)

    print(f"{num_images} images have been successfully created and saved to {output_dir}")

# Example usage
folders = ['E:/Coins/Augmented Coins/Thai Baht/1', 
           'E:/Coins/Augmented Coins/Thai Baht/5',
           'E:/Coins/Augmented Coins/Thai Baht/10',
           'E:/Coins/Augmented Coins/Philippine Peso/1',
           'E:/Coins/Augmented Coins/Philippine Peso/5',
           'E:/Coins/Augmented Coins/Philippine Peso/10',
           'E:/Coins/Augmented Coins/Korean Won/10',
           'E:/Coins/Augmented Coins/Korean Won/50',
           'E:/Coins/Augmented Coins/Korean Won/100',
           'E:/Coins/Augmented Coins/Korean Won/500',
           'E:/Coins/Augmented Coins/Japanese Yen/1',
           'E:/Coins/Augmented Coins/Japanese Yen/5',
           'E:/Coins/Augmented Coins/Japanese Yen/10',
           'E:/Coins/Augmented Coins/Japanese Yen/50',
           'E:/Coins/Augmented Coins/Japanese Yen/100',
           'E:/Coins/Augmented Coins/Japanese Yen/500',
           'E:/Coins/Augmented Coins/Chinese Yuan',]  # Add the paths of your coin folders
output_dir = 'E:/Coins/Combined Coins'  # Directory to save the output images
num_images = 1500  # Number of images to generate

combine_random_coins_from_folders(folders, output_dir, num_images=num_images)
```
